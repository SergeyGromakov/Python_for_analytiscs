##### Условие 3: 3 задача
###### Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
###### Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
import seaborn as sns
sns.jointplot(x = house['sqft_living'], y = house['price'], kind = 'reg')
plt.title('Зависимость стоимости жилья от квадратуры')
plt.xlabel('Площадь')
plt.ylabel('Стоимость')
```

в целом наблюдается прямая линейная зависимость - выше квадратура - выше стоимость, однако в премиальном сегменте разброс цен очень высокий, что говорит о других специфичных факторах влияния

```python
house_cond = house.groupby('condition').agg({'price':'mean'}).reset_index()
plt.plot(house_cond['condition'], house_cond['price'])
plt.title('Зависимость стоимости жилья от состояния дома')
plt.xlabel('Состояние')
plt.ylabel('Стоимость')
```

в целом мы видим, что все дорогие дома в хорошем или отличном состоянии 

```python
house_badr = house.groupby('bedrooms').agg({'price':'mean'}).sort_index().reset_index()
plt.bar(house_badr['bedrooms'], house_badr['price'])
plt.title('Зависимость стоимости жилья от количества спален')
plt.xlabel('Количество спален')
plt.ylabel('Стоимость')
```

Дома с количеством  0 спален очевидно студии, которые популярны у зажиточных людей. Далее мы видим устойчивый рост кол-ва спален, а затем снижение. Очевиднео это связано, что дома с большим кол-вом спален являются общежитиями. 

```python
house_bath = house.groupby('bathrooms').agg({'price':'mean'}).sort_index().reset_index()
plt.scatter(house_bath['bathrooms'], house_bath['price'])
plt.title('Зависимость стоимости жилья от количества ванн')
plt.xlabel('Количество ванн')
plt.ylabel('Стоимость')
```

Видим практически прямой рост цены от кол-ва ванных, с небольшими выбросами

```python
plt.figure(figsize = (20, 10))
house_year = house.groupby('yr_built').agg({'price':'mean'}).sort_index().reset_index()
sns.barplot(x = house_year['yr_built'], y = house_year['price'])
plt.title('Зависимость стоимости жилья от года постройки')
plt.xlabel('Год постройки')
plt.ylabel('Стоимость')
plt.xticks(rotation = -90)
```

В целом видим отсутсвие зависимости стоимости от года постройки
