1.1 Изучите количество памяти с помощью matplotlib


```python
import matplotlib.pyplot as plt
plt.figure(figsize = (20, 4))
plt.hist(df['Memory_Amount'])
plt.title('График')
plt.xlabel('Размер памяти')
plt.ylabel('Количество')
plt.xticks(rotation = -90)
plt.yticks(rotation = 90)
```

3.1 Постройте график в абсолютных величинах


```python
df.head(5)
```

```python
data = pd.crosstab(index= df['Company'], columns= df['Cpu_Company'])
data = data.drop(columns = ['Samsung']).reset_index()
data
```

```python
import numpy as np
n_tick = np.arange(len(data['AMD']))
offset = 0.2
w = 0.4
plt.bar(n_tick - offset, data['AMD'], width = w)
plt.bar(n_tick + offset, data['Intel'], width = w)
plt.title('График распределения')
plt.xlabel('Компании')
plt.ylabel('Количество')
plt.legend(['AMD', 'Intel'])
plt.xticks(n_tick, data['Company'], rotation = 90)
```
